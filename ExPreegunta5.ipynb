Usando los métodos estudiados, determine todas las raíces de la ecuación:

$x^5+5.25\cdot x^4+4.125\cdot x^3-9.125\cdot x^2-14.625\cdot x-5.625=0$

```python
import numpy as np
from scipy import optimize

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

# Grilla de puntos iniciales para Newton
x0_values = np.linspace(-5, 2, 50)

raices = []
for x0 in x0_values:
    try:
        r = optimize.newton(f, x0, tol=1e-12, maxiter=100)
        # Guardar solo raíces únicas (tolerancia numérica)
        if not any(abs(r - r2) < 1e-8 for r2 in raices):
            raices.append(r)
    except (RuntimeError, OverflowError):
        pass

raices.sort()
for r in raices:
    print(f"{r:.12f}")
```
